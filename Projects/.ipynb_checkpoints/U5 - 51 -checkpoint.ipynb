## Agregación de datos por categoría

```python
import pandas as pd
import numpy as np
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500

gender_data = []
income_data = []

for i in range(0, n):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[1:10]
```

```python
height = 160 + 30 * np.random.randn(n)
weight = 65 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3500 * np.random.randn(n)
```

```python
data = pd.DataFrame(
    {
        "Gender": gender_data,
        "Economic Status": income_data,
        "Height": height,
        "Weight": weight,
        "Age": age,
        "Income": income
    }
)
```

```python
data.head()
```

## Agrupación de datos 

```python
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups #indica los valores y las posiciones 
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group("Female")
```

```python
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

# Operaciones sobre datos agrupados

```python
double_group.sum()
```

```python
double_group.mean() #media
```

```python
double_group.size() #el tamaño de los grupos
```

```python
double_group.describe() #con la función describe se obtienen los estadísticos promedios
```

```python
double_group.aggregate(
    {
        "Income": np.sum,
        "Age": np.mean,
        "Height": np.std
    }
)
```

## Filtrado de Datos

```python
double_group["Age"].filter(lambda x: x.sum()>2400)
```
